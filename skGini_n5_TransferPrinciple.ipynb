```python
import numpy as np
from numpy import random
import pandas as pd
import math
import statistics as stats
import scipy.stats 
from scipy.stats import skew
```

```python
dataframe2 = pd.DataFrame(columns=['TransferNr','Quantiles','TransferSum','CV',
                                       'Gini','Ginitf','Ginid',
                                       'rGini','rGinitf','rGinid',
                                       'lGini','lGinitf','lGinid',
                                       'Gsk','Gsktf','Gskd',
                                       'sGinil','deltal','sGinir','deltar',
                                       'skewness_paerson1','Skew','Skew2',
                                       'dMetric','absdMetric','Summ','dMetricDollar','bMeasure',
                                       'skewdivergence','SkewMMconform','bMMconform',
                                       'mean','median','stdv',
                                       'percent1','percent2','percent3','percent4','percent5',
                                       'percent1tf','percent2tf','percent3tf','percent4tf','percent5tf',
                                       'value1','value2','value3','value4','value5',
                                       'value1tf','value2tf','value3tf','value4tf','value5tf'        
                                  ])
```

```python
i=0
while i <=5000:
   
    value1=np.random.randint(1,50)
    value2=np.random.randint(1,100)
    value3=np.random.randint(1,150)
    value4=np.random.randint(1,300)
    value5=np.random.randint(2,500)
  
    wlist =[value1,value2,value3,value4,value5]
    wlist.sort()
    a=wlist[0]
    b=wlist[1]
    c=wlist[2]
    d=wlist[3]
    e=wlist[4]
    
    distr =([value1,value2,value3,value4,value5])
    Summ=(a+b+c+d+e)
    mean=(a+b+c+d+e)/5
    median=np.median(wlist)
    stdv=np.std(wlist)
    #Coefficient of Variation
    CV=stdv/mean
    
    ################################
    #lGini weights
    #4+3+2+1=10
    #10/4=2.5
    lweight1=4/2.5
    lweight2=3/2.5
    lweight3=2/2.5
    lweight4=1/2.5
    
    ##############################
    #rGini weights
    rweight1=1/2.5
    rweight2=2/2.5
    rweight3=3/2.5
    rweight4=4/2.5
  
    #####################################################################################################
    #################################################
    #Transfer                                                
    TransferNr=np.random.randint(1,10)
    TransferSum=random.choice([0.1,0.2,0.3,0.4,0.45,0.49])
    
    ###############################
    
    if(TransferNr==1):
        value1tf=a+TransferSum
        value2tf=b-TransferSum
        value3tf=c
        value4tf=d
        value5tf=e
      
        Quantiles='b2a'
        Difference=abs(a-b)
        
    elif(TransferNr==2):
        value1tf=a+TransferSum
        value2tf=b
        value3tf=c-TransferSum
        value4tf=d
        value5tf=e
       
        Quantiles='c2a' 
        Difference=abs(a-c)
           
    elif(TransferNr==3):
        value1tf=a+TransferSum
        value2tf=b
        value3tf=c
        value4tf=d-TransferSum
        value5tf=e
     
        Quantiles='d2a'
        Difference=abs(a-d)
         
    elif(TransferNr==4):
        value1tf=a+TransferSum
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e-TransferSum
     
        Quantiles='e2a'
        Difference=abs(a-e)

    ################################
    elif(TransferNr==5):
        value1tf=a
        value2tf=b+TransferSum
        value3tf=c-TransferSum
        value4tf=d
        value5tf=e
    
        Quantiles='c2b'
        Difference=abs(b-c)
                           
    elif(TransferNr==6):
        value1tf=a
        value2tf=b+TransferSum
        value3tf=c
        value4tf=d-TransferSum
        value5tf=e
       
        Quantiles='d2b'
        Difference=abs(b-d)
          
    elif(TransferNr==7):
        value1tf=a
        value2tf=b+TransferSum
        value3tf=c
        value4tf=d
        value5tf=e-TransferSum
       
        Quantiles='e2b'
        Difference=abs(b-e)
       
    ###################################
    
    elif(TransferNr==8):
        value1tf=a
        value2tf=b
        value3tf=c+TransferSum
        value4tf=d-TransferSum
        value5tf=e
       
        Quantiles='d2c'
        Difference=abs(c-d)
          
    elif(TransferNr==9):
        value1tf=a
        value2tf=b
        value3tf=c+TransferSum
        value4tf=d
        value5tf=e-TransferSum

        Quantiles='e2c'
        Difference=abs(c-e)
         
    ##############################
    
    elif(TransferNr==10):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d+TransferSum
        value5tf=e-TransferSum
    
        Quantiles='e2d'
        Difference=abs(d-e)
        
    wlisttf =[value1tf,value2tf,value3tf,value4tf,value5tf]
    wlisttf.sort()
    
    atf=wlisttf[0]
    btf=wlisttf[1]
    ctf=wlisttf[2]
    dtf=wlisttf[3]
    etf=wlisttf[4]
  
    ###################################################################################################
    
    g1 = (a/(a+b+c+d+e))
    g2 = (b/(a+b+c+d+e))
    g3 = (c/(a+b+c+d+e))
    g4 = (d/(a+b+c+d+e))
    g5 = (e/(a+b+c+d+e))
    
    g1tf = (atf/(atf+btf+ctf+dtf+etf))
    g2tf = (btf/(atf+btf+ctf+dtf+etf))
    g3tf = (ctf/(atf+btf+ctf+dtf+etf))
    g4tf = (dtf/(atf+btf+ctf+dtf+etf))
    g5tf = (etf/(atf+btf+ctf+dtf+etf))
  
    ##############################################################################################
    #Income shares
    percent1=round(g1*100,50)
    percent2=round(g2*100,50)
    percent3=round(g3*100,50)
    percent4=round(g4*100,50)
    percent5=round(g5*100,50)
   
    PerSumme=(g1+g2+g3+g4+g5)*100
    
    percent1tf=round(g1tf*100,50)
    percent2tf=round(g2tf*100,50)
    percent3tf=round(g3tf*100,50)
    percent4tf=round(g4tf*100,50)
    percent5tf=round(g5tf*100,50)
  
    #Cummulative income shares
    q1cmltv=percent1
    q2cmltv=q1cmltv+percent2
    q3cmltv=q2cmltv+percent3
    q4cmltv=q3cmltv+percent4
    q5cmltv=q4cmltv+percent5
  
    q1cmltvtf=percent1tf
    q2cmltvtf=q1cmltvtf+percent2tf
    q3cmltvtf=q2cmltvtf+percent3tf
    q4cmltvtf=q3cmltvtf+percent4tf
    q5cmltvtf=q4cmltvtf+percent5tf
   
    ################################################################################################
    ###############################################################################################
    #Gini from the n-1 distances between Lorenz curve and 45-degree line
    
    quantile1=100/5 
    quantile2=100/5*2
    quantile3=100/5*3
    quantile4=100/5*4
   
    sumquantiles=(100/5*10)#(100/n)*(n*(n-1))/2
    
    #n-1 distances
    distance1=abs((quantile1)-(q1cmltv))
    distance2=abs((quantile2)-(q2cmltv))
    distance3=abs((quantile3)-(q3cmltv))
    distance4=abs((quantile4)-(q4cmltv))
  
    sumdistances=(distance1+distance2+distance3+distance4)
    
    preGini=sumdistances/sumquantiles
    
    Gini=preGini*(4/5)
    
    ###############################################################################################
    #Gini from n-1 distances after the progressive transfer

    distance1tf=abs((quantile1)-(q1cmltvtf))
    distance2tf=abs((quantile2)-(q2cmltvtf))
    distance3tf=abs((quantile3)-(q3cmltvtf))
    distance4tf=abs((quantile4)-(q4cmltvtf))
   
    sumdistancestf=(distance1tf+distance2tf+distance3tf+distance4tf)
    
    preGinitf=sumdistancestf/sumquantiles
    
    Ginitf=preGinitf*(4/5)
    Ginid=100-((100/Gini)*Ginitf)
    
    ################################################################################################
    ###############################################################################################
    #G_sk+
    
    rdistance1=abs((quantile1)-(q1cmltv))*rweight1
    rdistance2=abs((quantile2)-(q2cmltv))*rweight2
    rdistance3=abs((quantile3)-(q3cmltv))*rweight3
    rdistance4=abs((quantile4)-(q4cmltv))*rweight4
  
    sumrdistances=(rdistance1+rdistance2+rdistance3+rdistance4)
    
    prerGini=sumrdistances/sumquantiles
    
    rGini=prerGini*(4/5)
    
    ###############################################################################################
    #G_sk+ after the progressive transfer
    
    rdistance1tf=abs((quantile1)-(q1cmltvtf))*rweight1
    rdistance2tf=abs((quantile2)-(q2cmltvtf))*rweight2
    rdistance3tf=abs((quantile3)-(q3cmltvtf))*rweight3
    rdistance4tf=abs((quantile4)-(q4cmltvtf))*rweight4
   
    sumrdistancestf=(rdistance1tf+rdistance2tf+rdistance3tf+rdistance4tf)
    
    prerGinitf=sumrdistancestf/sumquantiles
    
    rGinitf=prerGinitf*(4/5)
    
    rGinid=100-((100/rGini)*rGinitf)
    
    ################################################################################################
    ###############################################################################################
    #G_sk-
    
    ldistance1=abs((quantile1)-(q1cmltv))*lweight1
    ldistance2=abs((quantile2)-(q2cmltv))*lweight2
    ldistance3=abs((quantile3)-(q3cmltv))*lweight3
    ldistance4=abs((quantile4)-(q4cmltv))*lweight4
  
    sumldistances=(ldistance1+ldistance2+ldistance3+ldistance4)
    
    prelGini=sumldistances/sumquantiles
    
    lGini=prelGini*(4/5)
    
    ###############################################################################################
    #G_sk- after the progressive transfer
    
    ldistance1tf=abs((quantile1)-(q1cmltvtf))*lweight1
    ldistance2tf=abs((quantile2)-(q2cmltvtf))*lweight2
    ldistance3tf=abs((quantile3)-(q3cmltvtf))*lweight3
    ldistance4tf=abs((quantile4)-(q4cmltvtf))*lweight4
    
    sumldistancestf=(ldistance1tf+ldistance2tf+ldistance3tf+ldistance4tf)
    
    prelGinitf=sumldistancestf/sumquantiles
    
    lGinitf=prelGinitf*(4/5)
    
    
    lGinid=100-((100/lGini)*lGinitf)
    ##########################################################
    #Skweness adjusted Gini coefficient
    Gsk=Gini+(abs(rGini-lGini)/2)
    Gsktf=Ginitf+(abs(rGinitf-lGinitf)/2)
    Gskd=100-((100/Gsk)*Gsktf)
   
    ######################################################################
    #sGini sensitive to right skewness
    deltar=1.5
    N5=5
    
    x1r=(N5-1+1)
    x2r=(N5-2+1)
    x3r=(N5-3+1)
    x4r=(N5-4+1)
    x5r=(N5-5+1)
   
    y1r=(N5-1)
    y2r=(N5-2)
    y3r=(N5-3)
    y4r=(N5-4)
    y5r=(N5-5)
  
    diff1r=pow(x1r,deltar)-pow(y1r,deltar)
    diff2r=pow(x2r,deltar)-pow(y2r,deltar)
    diff3r=pow(x3r,deltar)-pow(y3r,deltar)
    diff4r=pow(x4r,deltar)-pow(y4r,deltar)
    diff5r=pow(x5r,deltar)-pow(y5r,deltar)
     
    sGiniar=(a/mean)*diff1r
    sGinibr=(b/mean)*diff2r
    sGinicr=(c/mean)*diff3r
    sGinidr=(d/mean)*diff4r
    sGinier=(e/mean)*diff5r
    
    sGiniSumr=(sGiniar+sGinibr+sGinicr+sGinidr+sGinier)
    
    sGinir=1-(sGiniSumr*(1/pow(5,deltar)))
    #################################################################
    #sGini sensitive to left skewness
    deltal=3
    N5=5
    
    x1l=(N5-1+1)
    x2l=(N5-2+1)
    x3l=(N5-3+1)
    x4l=(N5-4+1)
    x5l=(N5-5+1)
    
    y1l=(N5-1)
    y2l=(N5-2)
    y3l=(N5-3)
    y4l=(N5-4)
    y5l=(N5-5)
    
    diff1l=pow(x1l,deltal)-pow(y1l,deltal)
    diff2l=pow(x2l,deltal)-pow(y2l,deltal)
    diff3l=pow(x3l,deltal)-pow(y3l,deltal)
    diff4l=pow(x4l,deltal)-pow(y4l,deltal)
    diff5l=pow(x5l,deltal)-pow(y5l,deltal)
    
    sGinial=(a/mean)*diff1l
    sGinibl=(b/mean)*diff2l
    sGinicl=(c/mean)*diff3l
    sGinidl=(d/mean)*diff4l
    sGiniel=(e/mean)*diff5l
   
    sGiniSuml=(sGinial+sGinibl+sGinicl+sGinidl+sGiniel)
    
    sGinil=1-(sGiniSuml*(1/pow(5,deltal)))
    
    ########################################################################
    ########################################################################
    #Skewness Measures
    #######################################################################
    #Conventional skewness measures
    
    #Skewness based on mean, median and standard deviation
    skewness_paerson1=(3 * (mean - median)) / stdv
    
    #Skew: Standard method to calculate skewness based on higher moments
    
    skew1=pow((a-mean),3)
    skew2=pow((b-mean),3)
    skew3=pow((c-mean),3)
    skew4=pow((d-mean),3)
    skew5=pow((e-mean),3)
   
    skewsum=(skew1+skew2+skew3+skew4+skew5)
    
    Skew=skewsum/ (5 * pow(stdv,3))
    #Skew_=scipy.stats.skew(wlist, axis=0, bias=True)
    
    #Another version of a standard moment-based measure of skewness
    Skew2=((5/((5-1)*(5-2)))*(skewsum/pow(stdv,3)))
    
    ######################################################## 
    #Novel skewness measures
    
    #d-metric = (n/n-1)[1.5(rGini-lGini)]
    dMetric=(5/4)*(1.5*(rGini-lGini))
    absdMetric=abs((5/4)*(1.5*(rGini-lGini)))
    dMetricDollar=absdMetric*Summ
    
    #measure b 
    bMeasureComp1=(1-((1/rGini)*lGini))
    bMeasureComp2=(((1/lGini)*rGini)-1)
    
    bMeasure=(bMeasureComp1+bMeasureComp2)/1.5

    #################################
    # Does the novel measure b indicate a different skewness than the standard moment-based measure of skewness?
    if((bMeasure<0) and (Skew>0)):
        skewdivergence="yes"
    elif((bMeasure>0) and (Skew<0)):   
         skewdivergence="yes"
    elif((bMeasure<=0) and (Skew<=0)):
         skewdivergence="no"
    elif((bMeasure>=0) and (Skew>=0)):
         skewdivergence="no"
             
    #################################
    #Testing how the novel skewness measure b and the standard measure conform to the rule of thumb...
    #...that the mean is right of the median under right skew, and left of the median under left skew
    
    if((mean>median) and (bMeasure>=0)):
        bMMconform="yes"
    elif((mean<median) and (bMeasure<=0)):
        bMMconform="yes" 
    elif((mean<median) and (bMeasure>=0)):   
        bMMconform="no"
    elif((mean>median) and (bMeasure<0)):
        bMMconform="no"    
    #################################
    
    if((mean>median) and (Skew>=0)):
        SkewMMconform="yes"
    elif((mean<median) and (Skew<=0)):
        SkewMMconform="yes" 
    elif((mean<median) and (Skew>=0)):   
        SkewMMconform="no"
    elif((mean>median) and (Skew<0)):
        SkewMMconform="no"      
    ################################################################################################
    ###############################################################################################
    ##############################################################################################
    
    value1=a
    value2=b
    value3=c
    value4=d
    value5=e
  
    value1tf=atf
    value2tf=btf
    value3tf=ctf
    value4tf=dtf
    value5tf=etf
   
    #################################################################################
    i+=1
   
       
    dataframe2.loc[len(dataframe2)] = (TransferNr,Quantiles,TransferSum,CV,
                                       Gini,Ginitf,Ginid,
                                       rGini,rGinitf,rGinid,
                                       lGini,lGinitf,lGinid,
                                       Gsk,Gsktf,Gskd,
                                       sGinil,deltal,sGinir,deltar,
                                       skewness_paerson1,Skew,Skew2,
                                       dMetric,absdMetric,Summ,dMetricDollar,bMeasure,
                                       skewdivergence,SkewMMconform,bMMconform,
                                       mean,median,stdv,
                                       percent1,percent2,percent3,percent4,percent5,
                                       percent1tf,percent2tf,percent3tf,percent4tf,percent5tf,
                                       value1,value2,value3,value4,value5,
                                       value1tf,value2tf,value3tf,value4tf,value5tf
                                      )                                               
                                                                          
```
