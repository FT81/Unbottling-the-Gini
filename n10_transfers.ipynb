```python
i=0
while i <=10000:
   
    value1=np.random.randint(1,250)
    value2=np.random.randint(1,250)
    value3=np.random.randint(1,250)
    value4=np.random.randint(1,250)
    value5=np.random.randint(1,250)
    value6=np.random.randint(1,250)
    value7=np.random.randint(1,250)
    value8=np.random.randint(1,250)
    value9=np.random.randint(1,250)
    value10=np.random.randint(1,250)
   
    wlist =[value1,value2,value3,value4,value5,value6,value7,value8,value9,value10]
    wlist.sort()
    
    a=wlist[0]
    b=wlist[1]
    c=wlist[2]
    d=wlist[3]
    e=wlist[4]
    f=wlist[5]
    g=wlist[6]
    h=wlist[7]
    j=wlist[8]
    k=wlist[9]
    
    distr =([value1,value2,value3,value4,value5,value6,value7,value8,value9,value10])
    
    Summ=(a+b+c+d+e+f+g+h+j+k)
    mean=(a+b+c+d+e+f+g+h+j+k)/10
    median=np.median(wlist)
    vari=np.var(wlist)
    stdv=np.std(wlist)  
    
    ################################
    #lGini weights
    
    lweight1=9/5
    lweight2=8/5
    lweight3=7/5
    lweight4=6/5
    lweight5=5/5
    lweight6=4/5
    lweight7=3/5
    lweight8=2/5
    lweight9=1/5
    
    ##############################
    #rGini weights
    
    rweight1=1/5
    rweight2=2/5
    rweight3=3/5
    rweight4=4/5
    rweight5=5/5
    rweight6=6/5
    rweight7=7/5
    rweight8=8/5
    rweight9=9/5
   
    #####################################################################################################
    ####################################################################################################
    #Transfers
    
    divisor=np.random.randint(3,100)

    TransferNr=np.random.randint(1,45)
   
    ###############################
    
    if(TransferNr==1):
        value1tf=a+(abs(a-b)/divisor)
        value2tf=b-(abs(a-b)/divisor)
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f 
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='b2a'
        Transfersumm=(abs(a-b)/divisor)
        RankDiff=1
        Difference=abs(a-b)
        
         
    elif(TransferNr==2):
        value1tf=a+(abs(a-c)/divisor)
        value2tf=b
        value3tf=c-(abs(a-c)/divisor)
        value4tf=d
        value5tf=e
        value6tf=f 
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='c2a' 
        Transfersumm=(abs(a-c)/divisor)
        RankDiff=2
        Difference=abs(a-c)
       
        
    elif(TransferNr==3):
        value1tf=a+(abs(a-d)/divisor)
        value2tf=b
        value3tf=c
        value4tf=d-(abs(a-d)/divisor)
        value5tf=e
        value6tf=f 
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='d2a'
        Transfersumm=(abs(a-d)/divisor)
        RankDiff=3
        Difference=abs(a-d)
       
        
    elif(TransferNr==4):
        value1tf=a+(abs(a-e)/divisor)
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e-(abs(a-e)/divisor)
        value6tf=f 
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='e2a'
        Transfersumm=(abs(a-e)/divisor)
        RankDiff=4
        Difference=abs(a-e)
       
        
    elif(TransferNr==5):
        value1tf=a+(abs(a-f)/divisor)
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f-(abs(a-f)/divisor)
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='f2a'   
        Transfersumm=(abs(a-f)/divisor)
        RankDiff=5
        Difference=abs(a-f)
       
        
    elif(TransferNr==6):
        value1tf=a+(abs(a-g)/divisor)
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f 
        value7tf=g-(abs(a-g)/divisor)
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='g2a'
        Transfersumm=(abs(a-g)/divisor) 
        RankDiff=6
        Difference=abs(a-g)
       
                       
    elif(TransferNr==7):
        value1tf=a+(abs(a-h)/divisor)
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f
        value7tf=g
        value8tf=h-(abs(a-h)/divisor)
        value9tf=j
        value10tf=k
        
        Quantiles='h2a'
        Transfersumm=(abs(a-h)/divisor)
        RankDiff=7
        Difference=abs(a-h)
        
                       
    elif(TransferNr==8):
        value1tf=a+(abs(a-j)/divisor)
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f 
        value7tf=g
        value8tf=h
        value9tf=j-(abs(a-j)/divisor)
        value10tf=k 
        
        Quantiles='j2a'
        Transfersumm=(abs(a-j)/divisor)  
        RankDiff=8
        Difference=abs(a-j)
       
    elif(TransferNr==9):
        value1tf=a+(abs(a-k)/divisor)
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f 
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k-(abs(a-k)/divisor)
        
        Quantiles='k2a'
        Transfersumm=(abs(a-k)/divisor)
        RankDiff=9
        Difference=abs(a-k)
        
                       
    ################################
    elif(TransferNr==10):
        value1tf=a
        value2tf=b+(abs(b-c)/divisor)
        value3tf=c-(abs(b-c)/divisor)
        value4tf=d
        value5tf=e
        value6tf=f 
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='c2b'
        Transfersumm=(abs(b-c)/divisor)
        RankDiff=1
        Difference=abs(b-c)
        
                       
    elif(TransferNr==11):
        value1tf=a
        value2tf=b+(abs(b-d)/divisor)
        value3tf=c
        value4tf=d-(abs(b-d)/divisor)
        value5tf=e
        value6tf=f 
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='d2b'
        Transfersumm=(abs(b-d)/divisor)
        RankDiff=2
        Difference=abs(b-d)
       
        
    elif(TransferNr==12):
        value1tf=a
        value2tf=b+(abs(b-e)/divisor)
        value3tf=c
        value4tf=d
        value5tf=e-(abs(b-e)/divisor)
        value6tf=f 
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='e2b'
        Transfersumm=(abs(b-e)/divisor) 
        RankDiff=3
        Difference=abs(b-e)
       
        
    elif(TransferNr==13):
        value1tf=a
        value2tf=b+(abs(b-f)/divisor)
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f-(abs(b-f)/divisor)
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='f2b'
        Transfersumm=(abs(b-f)/divisor)
        RankDiff=4
        Difference=abs(b-f)
        
    elif(TransferNr==14):
        value1tf=a
        value2tf=b+(abs(b-g)/divisor)
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f
        value7tf=g-(abs(b-g)/divisor)
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='g2b'
        Transfersumm=(abs(b-g)/divisor)
        RankDiff=5
        Difference=abs(b-g)
       
        
    elif(TransferNr==15):
        value1tf=a
        value2tf=b+(abs(b-h)/divisor)
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f 
        value7tf=g
        value8tf=h-(abs(b-h)/divisor)
        value9tf=j
        value10tf=k  
        
        Quantiles='h2b'
        Transfersumm=(abs(b-h)/divisor) 
        RankDiff=6
        Difference=abs(b-h) 
        
        
    elif(TransferNr==16):
        value1tf=a
        value2tf=b+(abs(b-j)/divisor)
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f
        value7tf=g
        value8tf=h
        value9tf=j-(abs(b-j)/divisor)
        value10tf=k
        
        Quantiles='j2b'
        Transfersumm=(abs(b-j)/divisor)
        RankDiff=7
        Difference=abs(b-j)
       
        
    elif(TransferNr==17):
        value1tf=a
        value2tf=b+(abs(b-k)/divisor)
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f 
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k-(abs(b-k)/divisor)
        
        Quantiles='k2b'
        Transfersumm=(abs(b-k)/divisor)
        RankDiff=8
        Difference=abs(b-k)
        
    ###################################
    
    elif(TransferNr==18):
        value1tf=a
        value2tf=b
        value3tf=c+(abs(c-d)/divisor)
        value4tf=d-(abs(c-d)/divisor)
        value5tf=e
        value6tf=f 
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='d2c'
        Transfersumm=(abs(c-d)/divisor) 
        RankDiff=1
        Difference=abs(c-d)
       
        
    elif(TransferNr==19):
        value1tf=a
        value2tf=b
        value3tf=c+(abs(c-e)/divisor)
        value4tf=d
        value5tf=e-(abs(c-e)/divisor)
        value6tf=f 
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='e2c'
        Transfersumm=(abs(c-e)/divisor) 
        RankDiff=2
        Difference=abs(c-e)
       
        
    elif(TransferNr==20):
        value1tf=a
        value2tf=b
        value3tf=c+(abs(c-f)/divisor)
        value4tf=d
        value5tf=e
        value6tf=f-(abs(c-f)/divisor)
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='f2c'
        Transfersumm=(abs(c-f)/divisor) 
        RankDiff=3
        Difference=abs(c-f)
        
        
    elif(TransferNr==21):
        value1tf=a
        value2tf=b
        value3tf=c+(abs(c-g)/divisor)
        value4tf=d
        value5tf=e
        value6tf=f
        value7tf=g-(abs(c-g)/divisor)
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='g2c'
        Transfersumm=(abs(c-g)/divisor)
        RankDiff=4
        Difference=abs(c-g)
       
        
    elif(TransferNr==22):
        value1tf=a
        value2tf=b
        value3tf=c+(abs(c-h)/divisor)
        value4tf=d
        value5tf=e
        value6tf=f 
        value7tf=g
        value8tf=h-(abs(c-h)/divisor)
        value9tf=j
        value10tf=k  
        
        Quantiles='h2c'
        Transfersumm=(abs(c-h)/divisor)
        RankDiff=5
        Difference=abs(c-h)
        
        
    elif(TransferNr==23):
        value1tf=a
        value2tf=b
        value3tf=c+(abs(c-j)/divisor)
        value4tf=d
        value5tf=e
        value6tf=f
        value7tf=g
        value8tf=h
        value9tf=j-(abs(c-j)/divisor)
        value10tf=k
        
        Quantiles='j2c'
        Transfersumm=(abs(c-j)/divisor)
        RankDiff=6
        Difference=abs(c-j)
        
        
    elif(TransferNr==24):
        value1tf=a
        value2tf=b
        value3tf=c+(abs(c-k)/divisor)
        value4tf=d
        value5tf=e
        value6tf=f 
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k-(abs(c-k)/divisor)
        
        Quantiles='k2c'
        Transfersumm=(abs(c-k)/divisor)
        RankDiff=7
        Difference=abs(c-k)
        
        
    ##############################
    
    elif(TransferNr==25):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d+(abs(d-e)/divisor)
        value5tf=e-(abs(d-e)/divisor)
        value6tf=f 
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='e2d'
        Transfersumm=(abs(e-d)/divisor)
        RankDiff=1
        Difference=abs(d-e)
        
        
    elif(TransferNr==26):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d+(abs(d-f)/divisor)
        value5tf=e
        value6tf=f-(abs(d-f)/divisor)
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='f2d'
        Transfersumm=(abs(f-d)/divisor) 
        RankDiff=2
        Difference=abs(d-f)
        
        
    elif(TransferNr==27):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d+(abs(d-g)/divisor)
        value5tf=e
        value6tf=f
        value7tf=g-(abs(d-g)/divisor)
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='g2d'
        Transfersumm=(abs(g-d)/divisor) 
        RankDiff=3
        Difference=abs(d-g)
        
    elif(TransferNr==28):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d+(abs(d-h)/divisor)
        value5tf=e
        value6tf=f 
        value7tf=g
        value8tf=h-(abs(d-h)/divisor)
        value9tf=j
        value10tf=k  
        
        Quantiles='h2d'
        Transfersumm=(abs(h-d)/divisor)  
        RankDiff=4
        Difference=abs(d-h)
       
        
    elif(TransferNr==29):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d+(abs(d-j)/divisor)
        value5tf=e
        value6tf=f
        value7tf=g
        value8tf=h
        value9tf=j-(abs(d-j)/divisor)
        value10tf=k
        
        Quantiles='j2d'
        Transfersumm=(abs(j-d)/divisor) 
        RankDiff=5
        Difference=abs(d-j)
       
        
    elif(TransferNr==30):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d+(abs(d-k)/divisor)
        value5tf=e
        value6tf=f 
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k-(abs(d-k)/divisor)
        
        Quantiles='k2d'
        Transfersumm=(abs(k-d)/divisor)
        RankDiff=6
        Difference=abs(d-k)
       
        
    #################################
    
    elif(TransferNr==31):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e+(abs(e-f)/divisor)
        value6tf=f-(abs(e-f)/divisor)
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='f2e'
        Transfersumm=(abs(f-e)/divisor)
        RankDiff=1
        Difference=abs(e-f)
       
        
    elif(TransferNr==32):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e+(abs(e-g)/divisor)
        value6tf=f
        value7tf=g-(abs(e-g)/divisor)
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='g2e'
        Transfersumm=(abs(g-e)/divisor) 
        RankDiff=2
        Difference=abs(e-g)
       
        
    elif(TransferNr==33):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e+(abs(e-h)/divisor)
        value6tf=f 
        value7tf=g
        value8tf=h-(abs(e-h)/divisor)
        value9tf=j
        value10tf=k  
        
        Quantiles='h2e'
        Transfersumm=(abs(h-e)/divisor) 
        RankDiff=3
        Difference=abs(e-h)
       
        
    elif(TransferNr==34):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e+(abs(e-j)/divisor)
        value6tf=f
        value7tf=g
        value8tf=h
        value9tf=j-(abs(e-j)/divisor)
        value10tf=k
        
        Quantiles='j2e'
        Transfersumm=(abs(j-e)/divisor)
        RankDiff=4
        Difference=abs(e-j)
        
        
    elif(TransferNr==35):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e+(abs(e-k)/divisor)
        value6tf=f 
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k-(abs(e-k)/divisor)
        
        Quantiles='k2e'
        Transfersumm=(abs(k-e)/divisor) 
        RankDiff=5
        Difference=abs(e-k)
       

    ###########################
    
    elif(TransferNr==36):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f+(abs(f-g)/divisor)
        value7tf=g-(abs(f-g)/divisor)
        value8tf=h
        value9tf=j
        value10tf=k
        
        Quantiles='g2f'
        Transfersumm=(abs(g-f)/divisor) 
        RankDiff=1
        Difference=abs(f-g)
        
        
    elif(TransferNr==37):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f+(abs(h-f)/divisor)
        value7tf=g
        value8tf=h-(abs(h-f)/divisor)
        value9tf=j
        value10tf=k 
        
        Quantiles='h2f'
        Transfersumm=(abs(h-f)/divisor)
        RankDiff=2
        Difference=abs(f-h)
       
        
    elif(TransferNr==38):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f+(abs(j-f)/divisor)
        value7tf=g
        value8tf=h
        value9tf=j-(abs(j-f)/divisor)
        value10tf=k 
        
        Quantiles='j2f'
        Transfersumm=(abs(j-f)/divisor) 
        RankDiff=3
        Difference=abs(f-j)
        
        
    elif(TransferNr==39):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f+(abs(k-f)/divisor)
        value7tf=g
        value8tf=h
        value9tf=j
        value10tf=k-(abs(k-f)/divisor)
        
        Quantiles='k2f'
        Transfersumm=(abs(k-f)/divisor)
        RankDiff=4
        Difference=abs(f-k)
       
        
    ###########################
    
    elif(TransferNr==40):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f
        value7tf=g+(abs(g-h)/divisor)
        value8tf=h-(abs(g-h)/divisor)
        value9tf=j
        value10tf=k  
        
        Quantiles='h2g'
        Transfersumm=(abs(h-g)/divisor) 
        RankDiff=1
        Difference=abs(g-h)
        
        
    elif(TransferNr==41):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f
        value7tf=g+(abs(j-g)/divisor)
        value8tf=h
        value9tf=j-(abs(j-g)/divisor)
        value10tf=k
        
        Quantiles='j2g'
        Transfersumm=(abs(j-g)/divisor)
        RankDiff=2
        Difference=abs(g-j)
        
        
    elif(TransferNr==42):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f 
        value7tf=g+(abs(g-k)/divisor)
        value8tf=h
        value9tf=j
        value10tf=k-(abs(g-k)/divisor)
        
        Quantiles='k2g'
        Transfersumm=(abs(k-g)/divisor) 
        RankDiff=3
        Difference=abs(g-k)
        
        
    ########################
    
    elif(TransferNr==43):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f
        value7tf=g
        value8tf=h+(abs(h-j)/divisor) 
        value9tf=j-(abs(h-j)/divisor)
        value10tf=k
        
        Quantiles='j2h'
        Transfersumm=(abs(j-h)/divisor)  
        RankDiff=1
        Difference=abs(h-j)
        
        
    elif(TransferNr==44):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f
        value7tf=g
        value8tf=h+(abs(h-k)/divisor)
        value9tf=j
        value10tf=k-(abs(h-k)/divisor)
        
        Quantiles='k2h'
        Transfersumm=(abs(k-h)/divisor)
        RankDiff=2
        Difference=abs(h-k)
        
        
    #####################
    
    elif(TransferNr==45):
        value1tf=a
        value2tf=b
        value3tf=c
        value4tf=d
        value5tf=e
        value6tf=f
        value7tf=g
        value8tf=h
        value9tf=j+(abs(j-k)/divisor) 
        value10tf=k-(abs(j-k)/divisor)
        
        Quantiles='k2j'
        Transfersumm=(abs(j-k)/divisor) 
        RankDiff=1
        Difference=abs(j-k)
        
       
    #######################################################
    
   
    
    wlisttf =[value1tf,value2tf,value3tf,value4tf,value5tf,value6tf,value7tf,value8tf,value9tf,value10tf]
    wlisttf.sort()
    
    atf=wlisttf[0]
    btf=wlisttf[1]
    ctf=wlisttf[2]
    dtf=wlisttf[3]
    etf=wlisttf[4]
    ftf=wlisttf[5]
    gtf=wlisttf[6]
    htf=wlisttf[7]
    jtf=wlisttf[8]
    ktf=wlisttf[9]
    
   
    ###################################################################################################
    #cGini
    
    g1 = (a/(a+b+c+d+e+f+g+h+j+k))
    g2 = (b/(a+b+c+d+e+f+g+h+j+k))
    g3 = (c/(a+b+c+d+e+f+g+h+j+k))
    g4 = (d/(a+b+c+d+e+f+g+h+j+k))
    g5 = (e/(a+b+c+d+e+f+g+h+j+k))
    g6 = (f/(a+b+c+d+e+f+g+h+j+k))
    g7 = (g/(a+b+c+d+e+f+g+h+j+k))
    g8 = (h/(a+b+c+d+e+f+g+h+j+k))
    g9 = (j/(a+b+c+d+e+f+g+h+j+k))
    g10 = (k/(a+b+c+d+e+f+g+h+j+k))
   
    summ = (g1+g2+g3+g4+g5+g6+g7+g8+g9+g10)
    glist = [g1,g2,g3,g4,g5,g6,g7,g8,g9,g10]
    glist.sort()
    
    v1=glist[0]
    v2=glist[1]
    v3=glist[2]
    v4=glist[3]
    v5=glist[4]
    v6=glist[5]
    v7=glist[6]
    v8=glist[7]
    v9=glist[8]
    v10=glist[9]
    
    c1=glist[0]
    c2=v1+glist[1]
    c3=v2+glist[2]
    c4=v3+glist[3]
    c5=v4+glist[4]
    c6=v5+glist[5]
    c7=v6+glist[6]
    c8=v7+glist[7]
    c9=v8+glist[8]
    c10=v9+glist[9] 
   
    freq=1/10
    f1=freq
    f2=f1+freq
    f3=f2+freq
    f4=f3+freq
    f5=f4+freq
    f6=f5+freq
    f7=f6+freq
    f8=f7+freq
    f9=f8+freq
    f10=f9+freq
  
    gi1=f1*v1
    gi2=(f1+f2)*v2
    gi3=(f2+f3)*v3
    gi4=(f3+f4)*v4
    gi5=(f4+f5)*v5
    gi6=(f5+f6)*v6
    gi7=(f6+f7)*v7
    gi8=(f7+f8)*v8
    gi9=(f8+f9)*v9
    gi10=(f9+f10)*v10
   
    gisumm=(gi1+gi2+gi3+gi4+gi5+gi6+gi7+gi8+gi9+gi10)
    cGini=gisumm-1
    
  
    
    ###############################################################################################
    #cGini after a progressive transfer
    
    g1tf = (atf/(atf+btf+ctf+dtf+etf+ftf+gtf+htf+jtf+ktf))
    g2tf = (btf/(atf+btf+ctf+dtf+etf+ftf+gtf+htf+jtf+ktf))
    g3tf = (ctf/(atf+btf+ctf+dtf+etf+ftf+gtf+htf+jtf+ktf))
    g4tf = (dtf/(atf+btf+ctf+dtf+etf+ftf+gtf+htf+jtf+ktf))
    g5tf = (etf/(atf+btf+ctf+dtf+etf+ftf+gtf+htf+jtf+ktf))
    g6tf = (ftf/(atf+btf+ctf+dtf+etf+ftf+gtf+htf+jtf+ktf))
    g7tf = (gtf/(atf+btf+ctf+dtf+etf+ftf+gtf+htf+jtf+ktf))
    g8tf = (htf/(atf+btf+ctf+dtf+etf+ftf+gtf+htf+jtf+ktf))
    g9tf = (jtf/(atf+btf+ctf+dtf+etf+ftf+gtf+htf+jtf+ktf))
    g10tf = (ktf/(atf+btf+ctf+dtf+etf+ftf+gtf+htf+jtf+ktf))
   
    summtf = (g1tf+g2tf+g3tf+g4tf+g5tf+g6tf+g7tf+g8tf+g9tf+g10tf)
    glisttf = [g1tf,g2tf,g3tf,g4tf,g5tf,g6tf,g7tf,g8tf,g9tf,g10tf]
    glisttf.sort()
    
    v1tf=glisttf[0]
    v2tf=glisttf[1]
    v3tf=glisttf[2]
    v4tf=glisttf[3]
    v5tf=glisttf[4]
    v6tf=glisttf[5]
    v7tf=glisttf[6]
    v8tf=glisttf[7]
    v9tf=glisttf[8]
    v10tf=glisttf[9]
    
    c1tf=glisttf[0]
    c2tf=v1tf+glisttf[1]
    c3tf=v2tf+glisttf[2]
    c4tf=v3tf+glisttf[3]
    c5tf=v4tf+glisttf[4]
    c6tf=v5tf+glisttf[5]
    c7tf=v6tf+glisttf[6]
    c8tf=v7tf+glisttf[7]
    c9tf=v8tf+glisttf[8]
    c10tf=v9tf+glisttf[9] 
   
    freqtf=1/10
    f1tf=freqtf
    f2tf=f1tf+freqtf
    f3tf=f2tf+freqtf
    f4tf=f3tf+freqtf
    f5tf=f4tf+freqtf
    f6tf=f5tf+freqtf
    f7tf=f6tf+freqtf
    f8tf=f7tf+freqtf
    f9tf=f8tf+freqtf
    f10tf=f9tf+freqtf
  
    gi1tf=f1tf*v1tf
    gi2tf=(f1tf+f2tf)*v2tf
    gi3tf=(f2tf+f3tf)*v3tf
    gi4tf=(f3tf+f4tf)*v4tf
    gi5tf=(f4tf+f5tf)*v5tf
    gi6tf=(f5tf+f6tf)*v6tf
    gi7tf=(f6tf+f7tf)*v7tf
    gi8tf=(f7tf+f8tf)*v8tf
    gi9tf=(f8tf+f9tf)*v9tf
    gi10tf=(f9tf+f10tf)*v10tf
   
    gisummtf=(gi1tf+gi2tf+gi3tf+gi4tf+gi5tf+gi6tf+gi7tf+gi8tf+gi9tf+gi10tf)
    cGinitf=gisummtf-1
   
    cGinid=100-((100/cGini)*cGinitf)
    ####################################################################
    #Coefficient of Variation
    CV=stdv/mean
  
    ##############################################################################################
    #Percent shares of total income
    percent1=round(g1*100,99)
    percent2=round(g2*100,99)
    percent3=round(g3*100,99)
    percent4=round(g4*100,99)
    percent5=round(g5*100,99)
    percent6=round(g6*100,99)
    percent7=round(g7*100,99)
    percent8=round(g8*100,99)
    percent9=round(g9*100,99)
    percent10=round(g10*100,99)
    
    Persumm=(g1+g2+g3+g4+g5+g6+g7+g8+g9+g10)*100
    
    percent1tf=round(g1tf*100,99)
    percent2tf=round(g2tf*100,99)
    percent3tf=round(g3tf*100,99)
    percent4tf=round(g4tf*100,99)
    percent5tf=round(g5tf*100,99)
    percent6tf=round(g6tf*100,99)
    percent7tf=round(g7tf*100,99)
    percent8tf=round(g8tf*100,99)
    percent9tf=round(g9tf*100,99)
    percent10tf=round(g10tf*100,99)
    
    Persummtf=(g1tf+g2tf+g3tf+g4tf+g5tf+g6tf+g7tf+g8tf+g9tf+g10tf)*100
    #Cummulative shares of income
    q1cmltv=percent1
    q2cmltv=q1cmltv+percent2
    q3cmltv=q2cmltv+percent3
    q4cmltv=q3cmltv+percent4
    q5cmltv=q4cmltv+percent5
    q6cmltv=q5cmltv+percent6
    q7cmltv=q6cmltv+percent7
    q8cmltv=q7cmltv+percent8
    q9cmltv=q8cmltv+percent9
    q10cmltv=q9cmltv+percent10
    
    q1cmltvtf=percent1tf
    q2cmltvtf=q1cmltvtf+percent2tf
    q3cmltvtf=q2cmltvtf+percent3tf
    q4cmltvtf=q3cmltvtf+percent4tf
    q5cmltvtf=q4cmltvtf+percent5tf
    q6cmltvtf=q5cmltvtf+percent6tf
    q7cmltvtf=q6cmltvtf+percent7tf
    q8cmltvtf=q7cmltvtf+percent8tf
    q9cmltvtf=q8cmltvtf+percent9tf
    q10cmltvtf=q9cmltvtf+percent10tf
   
   
    #################################################################################################
    ################################################################################################
    ###############################################################################################
    #Gini calcualted from distances between Lorenz curve and 45-degree line
    
    quantile1=10
    quantile2=20
    quantile3=30
    quantile4=40
    quantile5=50
    quantile6=60
    quantile7=70
    quantile8=80
    quantile9=90
    
    sumQuantiles=450
    
    #n-1 distance Segmente
    distance1=abs((quantile1)-(q1cmltv))
    distance2=abs((quantile2)-(q2cmltv))
    distance3=abs((quantile3)-(q3cmltv))
    distance4=abs((quantile4)-(q4cmltv))
    distance5=abs((quantile5)-(q5cmltv))
    distance6=abs((quantile6)-(q6cmltv))
    distance7=abs((quantile7)-(q7cmltv))
    distance8=abs((quantile8)-(q8cmltv))
    distance9=abs((quantile9)-(q9cmltv))
   
    
    sumdistancen=(distance1+distance2+distance3+distance4+
                   distance5+distance6+distance7+distance8+distance9)
    
    preGini=sumdistancen/sumQuantiles
    
    Gini=preGini*(9/10)
    
    ###############################################################################################
    #Gini after progressive transfer
    
    
    distance1tf=abs((quantile1)-(q1cmltvtf))
    distance2tf=abs((quantile2)-(q2cmltvtf))
    distance3tf=abs((quantile3)-(q3cmltvtf))
    distance4tf=abs((quantile4)-(q4cmltvtf))
    distance5tf=abs((quantile5)-(q5cmltvtf))
    distance6tf=abs((quantile6)-(q6cmltvtf))
    distance7tf=abs((quantile7)-(q7cmltvtf))
    distance8tf=abs((quantile8)-(q8cmltvtf))
    distance9tf=abs((quantile9)-(q9cmltvtf))
   
    
    sumdistancentf=(distance1tf+distance2tf+distance3tf+distance4tf+distance5tf+
                      distance6tf+distance7tf+distance8tf+distance9tf)
    
    preGinitf=sumdistancentf/sumQuantiles
    
    Ginitf=preGinitf*(9/10)
    
    #Difference between Gini and Ginitf
    Ginid=100-((100/Gini)*Ginitf)
    
    ################################################################################################
    ###############################################################################################
    #rGini( Gini_sk+)
    
    rdistance1=abs((quantile1)-(q1cmltv))*rweight1
    rdistance2=abs((quantile2)-(q2cmltv))*rweight2
    rdistance3=abs((quantile3)-(q3cmltv))*rweight3
    rdistance4=abs((quantile4)-(q4cmltv))*rweight4
    rdistance5=abs((quantile5)-(q5cmltv))*rweight5
    rdistance6=abs((quantile6)-(q6cmltv))*rweight6
    rdistance7=abs((quantile7)-(q7cmltv))*rweight7
    rdistance8=abs((quantile8)-(q8cmltv))*rweight8
    rdistance9=abs((quantile9)-(q9cmltv))*rweight9
   
    
    sumrdistancen=(rdistance1+rdistance2+rdistance3+rdistance4+rdistance5+
                     rdistance6+rdistance7+rdistance8+rdistance9)
    
    prerGini=sumrdistancen/sumQuantiles
    
    rGini=prerGini*(9/10)
    
    ###############################################################################################
    #rGini after the progressive transfer
    
    rdistance1tf=abs((quantile1)-(q1cmltvtf))*rweight1
    rdistance2tf=abs((quantile2)-(q2cmltvtf))*rweight2
    rdistance3tf=abs((quantile3)-(q3cmltvtf))*rweight3
    rdistance4tf=abs((quantile4)-(q4cmltvtf))*rweight4
    rdistance5tf=abs((quantile5)-(q5cmltvtf))*rweight5
    rdistance6tf=abs((quantile6)-(q6cmltvtf))*rweight6
    rdistance7tf=abs((quantile7)-(q7cmltvtf))*rweight7
    rdistance8tf=abs((quantile8)-(q8cmltvtf))*rweight8
    rdistance9tf=abs((quantile9)-(q9cmltvtf))*rweight9
   
    
    sumrdistancentf=(rdistance1tf+rdistance2tf+rdistance3tf+rdistance4tf+rdistance5tf+
                       rdistance6tf+rdistance7tf+rdistance8tf+rdistance9tf)
    
    prerGinitf=sumrdistancentf/sumQuantiles
    
    rGinitf=prerGinitf*(9/10)
    #Difference between rGini and rGinitf
    rGinid=100-((100/rGini)*rGinitf)
    
    ################################################################################################
    ###############################################################################################
    #lGini(=G_sk-)
    
    ldistance1=abs((quantile1)-(q1cmltv))*lweight1
    ldistance2=abs((quantile2)-(q2cmltv))*lweight2
    ldistance3=abs((quantile3)-(q3cmltv))*lweight3
    ldistance4=abs((quantile4)-(q4cmltv))*lweight4
    ldistance5=abs((quantile5)-(q5cmltv))*lweight5
    ldistance6=abs((quantile6)-(q6cmltv))*lweight6
    ldistance7=abs((quantile7)-(q7cmltv))*lweight7
    ldistance8=abs((quantile8)-(q8cmltv))*lweight8
    ldistance9=abs((quantile9)-(q9cmltv))*lweight9
   
    
    sumldistancen=(ldistance1+ldistance2+ldistance3+ldistance4+ldistance5+
                     ldistance6+ldistance7+ldistance8+ldistance9)
    
    prelGini=sumldistancen/sumQuantiles
    
    lGini=prelGini*(9/10)
    
    ###############################################################################################
    #lGini after the progressive transfer
    
    ldistance1tf=abs((quantile1)-(q1cmltvtf))*lweight1
    ldistance2tf=abs((quantile2)-(q2cmltvtf))*lweight2
    ldistance3tf=abs((quantile3)-(q3cmltvtf))*lweight3
    ldistance4tf=abs((quantile4)-(q4cmltvtf))*lweight4
    ldistance5tf=abs((quantile5)-(q5cmltvtf))*lweight5
    ldistance6tf=abs((quantile6)-(q6cmltvtf))*lweight6
    ldistance7tf=abs((quantile7)-(q7cmltvtf))*lweight7
    ldistance8tf=abs((quantile8)-(q8cmltvtf))*lweight8
    ldistance9tf=abs((quantile9)-(q9cmltvtf))*lweight9
   
    
    sumldistancentf=(ldistance1tf+ldistance2tf+ldistance3tf+ldistance4tf+ldistance5tf+
                       ldistance6tf+ldistance7tf+ldistance8tf+ldistance9tf)
    
    prelGinitf=sumldistancentf/sumQuantiles
    
    lGinitf=prelGinitf*(9/10)
    
    ##########################
    #Difference between lGini and lGinitf
    lGinid=100-((100/lGini)*lGinitf)
    ##########################################################
    #A skweness adjusted Gini coefficient
    Gsk=Gini+(abs(rGini-lGini)/2)
    Gsktf=Ginitf+(abs(rGinitf-lGinitf)/2)
    Gskd=100-((100/Gsk)*Gsktf)
    ######################################################################
    #Index S
    
    absdifferential1=abs(rdistance1-ldistance1)
    absdifferential2=abs(rdistance2-ldistance2)
    absdifferential3=abs(rdistance3-ldistance3)
    absdifferential4=abs(rdistance4-ldistance4)  
    absdifferential5=abs(rdistance5-ldistance5)
    absdifferential6=abs(rdistance6-ldistance6)
    absdifferential7=abs(rdistance7-ldistance7)
    absdifferential8=abs(rdistance8-ldistance8) 
    absdifferential9=abs(rdistance9-ldistance9) 
    
    sumabsdiffernetials=(absdifferential1+absdifferential2+absdifferential3+absdifferential4+
                         absdifferential5+absdifferential6+absdifferential7+absdifferential8+
                         absdifferential9)
    preIndexS=sumabsdiffernetials/sumQuantiles
    IndexS=preIndexS*9/10
    ##############################
    #Index W
    IndexW=Gini+(IndexS/2)
    ######################################################################
    #Index s after the progressive transfer 
    
    absdifferential1tf=abs(rdistance1tf-ldistance1tf)
    absdifferential2tf=abs(rdistance2tf-ldistance2tf)
    absdifferential3tf=abs(rdistance3tf-ldistance3tf)
    absdifferential4tf=abs(rdistance4tf-ldistance4tf)  
    absdifferential5tf=abs(rdistance5tf-ldistance5tf)
    absdifferential6tf=abs(rdistance6tf-ldistance6tf)
    absdifferential7tf=abs(rdistance7tf-ldistance7tf)
    absdifferential8tf=abs(rdistance8tf-ldistance8tf) 
    absdifferential9tf=abs(rdistance9tf-ldistance9tf) 
    
    sumabsdiffernetialstf=(absdifferential1tf+absdifferential2tf+absdifferential3tf+absdifferential4tf+
                           absdifferential5tf+absdifferential6tf+absdifferential7tf+absdifferential8tf+
                           absdifferential9tf)
    
    preIndexStf=sumabsdiffernetialstf/sumQuantiles
    IndexStf=preIndexStf*9/10
    #################################
    #Difference between original Index S and Index S after transfer
    IndexSd=100-((100/IndexS)*IndexStf)
    #################################
    #Index W after the progressive transfer
    IndexWtf=Ginitf+(IndexStf/2)
    ################
    #Difference between original Index W and Index W after transfer
    IndexWd=100-((100/IndexW)*IndexWtf)
    
    #####################################################################
    #sGini or extended Gini measure, sensitive to right-skewness
    deltar=1.5
    N10=10
    
    x1r=(N10-1+1)
    x2r=(N10-2+1)
    x3r=(N10-3+1)
    x4r=(N10-4+1)
    x5r=(N10-5+1)
    x6r=(N10-6+1)
    x7r=(N10-7+1)
    x8r=(N10-8+1)
    x9r=(N10-9+1)
    x10r=(N10-10+1)
    
    
    y1r=(N10-1)
    y2r=(N10-2)
    y3r=(N10-3)
    y4r=(N10-4)
    y5r=(N10-5)
    y6r=(N10-6)
    y7r=(N10-7)
    y8r=(N10-8)
    y9r=(N10-9)
    y10r=(N10-10)

    diff1r=pow(x1r,deltar)-pow(y1r,deltar)
    diff2r=pow(x2r,deltar)-pow(y2r,deltar)
    diff3r=pow(x3r,deltar)-pow(y3r,deltar)
    diff4r=pow(x4r,deltar)-pow(y4r,deltar)
    diff5r=pow(x5r,deltar)-pow(y5r,deltar)
    diff6r=pow(x6r,deltar)-pow(y6r,deltar)
    diff7r=pow(x7r,deltar)-pow(y7r,deltar)
    diff8r=pow(x8r,deltar)-pow(y8r,deltar)
    diff9r=pow(x9r,deltar)-pow(y9r,deltar)
    diff10r=pow(x10r,deltar)-pow(y10r,deltar)
    
    sGiniar=(a/mean)*diff1r
    sGinibr=(b/mean)*diff2r
    sGinicr=(c/mean)*diff3r
    sGinidr=(d/mean)*diff4r
    sGinier=(e/mean)*diff5r
    sGinifr=(f/mean)*diff6r
    sGinigr=(g/mean)*diff7r
    sGinihr=(h/mean)*diff8r
    sGinijr=(j/mean)*diff9r
    sGinikr=(k/mean)*diff10r
    
    sGiniSumr=(sGiniar+sGinibr+sGinicr+sGinidr+sGinier+
               sGinifr+sGinigr+sGinihr+sGinijr+sGinikr)
    
    sGinir=1-(sGiniSumr*(1/pow(10,deltar)))
    #################################################################
    #sGini or extended Gini measure sensitive to left skewness, full correlation with lGini(=G_sk-)
    deltal=3
    N10=10
    
    x1l=(N10-1+1)
    x2l=(N10-2+1)
    x3l=(N10-3+1)
    x4l=(N10-4+1)
    x5l=(N10-5+1)
    x6l=(N10-6+1)
    x7l=(N10-7+1)
    x8l=(N10-8+1)
    x9l=(N10-9+1)
    x10l=(N10-10+1)
    
    
    y1l=(N10-1)
    y2l=(N10-2)
    y3l=(N10-3)
    y4l=(N10-4)
    y5l=(N10-5)
    y6l=(N10-6)
    y7l=(N10-7)
    y8l=(N10-8)
    y9l=(N10-9)
    y10l=(N10-10)
 
    diff1l=pow(x1l,deltal)-pow(y1l,deltal)
    diff2l=pow(x2l,deltal)-pow(y2l,deltal)
    diff3l=pow(x3l,deltal)-pow(y3l,deltal)
    diff4l=pow(x4l,deltal)-pow(y4l,deltal)
    diff5l=pow(x5l,deltal)-pow(y5l,deltal)
    diff6l=pow(x6l,deltal)-pow(y6l,deltal)
    diff7l=pow(x7l,deltal)-pow(y7l,deltal)
    diff8l=pow(x8l,deltal)-pow(y8l,deltal)
    diff9l=pow(x9l,deltal)-pow(y9l,deltal)
    diff10l=pow(x10l,deltal)-pow(y10l,deltal)
    
    sGinial=(a/mean)*diff1l
    sGinibl=(b/mean)*diff2l
    sGinicl=(c/mean)*diff3l
    sGinidl=(d/mean)*diff4l
    sGiniel=(e/mean)*diff5l
    sGinifl=(f/mean)*diff6l
    sGinigl=(g/mean)*diff7l
    sGinihl=(h/mean)*diff8l
    sGinijl=(j/mean)*diff9l
    sGinikl=(k/mean)*diff10l
    
    sGiniSuml=(sGinial+sGinibl+sGinicl+sGinidl+sGiniel+
              sGinifl+sGinigl+sGinihl+sGinijl+sGinikl)
    
    sGinil=1-(sGiniSuml*(1/pow(10,deltal)))
    
   
    ########################################################################
    #Skewness Measures
    ########################################################################
    #Conventional skewness measures
    
    #Skewness based on mean, median and standard deviation
    skewness_paerson1=(3 * (mean - median)) / stdv
    
    #Skew: Standard method to calculate skewness based on higher moments
    
    skew1=pow((a-mean),3)
    skew2=pow((b-mean),3)
    skew3=pow((c-mean),3)
    skew4=pow((d-mean),3)
    skew5=pow((e-mean),3)
    skew6=pow((f-mean),3)
    skew7=pow((g-mean),3)
    skew8=pow((h-mean),3)
    skew9=pow((j-mean),3)
    skew10=pow((k-mean),3)
    
    skewsum=(skew1+skew2+skew3+skew4+skew5+
             skew6+skew7+skew8+skew9+skew10)
    
    Skew=skewsum/ (10 * pow(stdv,3))
    #Skew_=scipy.stats.skew(wlist, axis=0, bias=True)
    
    #Another version of a standard moment-based measure of skewness
    Skew2=((10/((10-1)*(10-2)))*(skewsum/pow(stdv,3)))
    
    ######################################################## 
    #Novel skewness measures
    
    #d-metric = (n/n-1)[1.5(rGini-lGini)]
    #Gsk equals G plus a third of the percentage expressed in the d-metric, for n large enough
    dMetric=(10/9)*(1.5*(rGini-lGini))
    absdMetric=abs((10/9)*(1.5*(rGini-lGini)))
    dMetricDollar=absdMetric*Summ
    
    #measure b 
    bMeasureComp1=(1-((1/rGini)*lGini))
    bMeasureComp2=(((1/lGini)*rGini)-1)
    
    bMeasure=(bMeasureComp1+bMeasureComp2)/1.5

    
    #################################
    # Does the novel measure b indicate a different skewness than the standard moment-based measure of skewness?
    if((bMeasure<0) and (Skew>0)):
        skewdivergence="yes"
    elif((bMeasure>0) and (Skew<0)):   
         skewdivergence="yes"
    elif((bMeasure<=0) and (Skew<=0)):
         skewdivergence="no"
    elif((bMeasure>=0) and (Skew>=0)):
         skewdivergence="no"
             
    #################################
    #Testing conformity to the rule of thumb, that the mean is right of the median under right skew, and left of the median under left skew
    
    if((mean>median) and (bMeasure>=0)):
        bMMconform="yes"
    elif((mean<median) and (bMeasure<=0)):
        bMMconform="yes" 
    elif((mean<median) and (bMeasure>=0)):   
        bMMconform="no"
    elif((mean>median) and (bMeasure<0)):
        bMMconform="no"    
    #################################
    
    if((mean>median) and (Skew>=0)):
        SkewMMconform="yes"
    elif((mean<median) and (Skew<=0)):
        SkewMMconform="yes" 
    elif((mean<median) and (Skew>=0)):   
        SkewMMconform="no"
    elif((mean>median) and (Skew<0)):
        SkewMMconform="no"        
    
    ################################################################################################
    ###############################################################################################
    ##############################################################################################
    
    value1=a
    value2=b
    value3=c
    value4=d
    value5=e
    value6=f
    value7=g
    value8=h
    value9=j
    value10=k
    
    value1tf=atf
    value2tf=btf
    value3tf=ctf
    value4tf=dtf
    value5tf=etf
    value6tf=ftf
    value7tf=gtf
    value8tf=htf
    value9tf=jtf
    value10tf=ktf

    #################################################################################
    i+=1
   
    dataframe2.loc[len(dataframe2)] = (TransferNr,Quantiles,CV,
                                       cGini,cGinitf,cGinid,
                                       Gini,Ginitf,Ginid,
                                       rGini,rGinitf,rGinid,
                                       lGini,lGinitf,lGinid,
                                       Gsk,Gsktf,Gskd,
                                       sGinil,deltal,sGinir,deltar,
                                       IndexS,IndexStf,IndexSd,
                                       IndexW,IndexWtf,IndexWd,
                                       skewness_paerson1,Skew,Skew2,
                                       dMetric,absdMetric,Summ,dMetricDollar,bMeasure,
                                       skewdivergence,SkewMMconform,bMMconform,
                                       mean,median,stdv,
                                       percent1,percent2,percent3,percent4,percent5,
                                       percent6,percent7,percent8,percent9,percent10,
                                       percent1tf,percent2tf,percent3tf,percent4tf,percent5tf,
                                       percent6tf,percent7tf,percent8tf,percent9tf,percent10tf,
                                       value1,value2,value3,value4,value5,
                                       value6,value7,value8,value9,value10,
                                       value1tf,value2tf,value3tf,value4tf,value5tf,
                                       value6tf,value7tf,value8tf,value9tf,value10tf            
                                      )                                               
                                                                          
```
